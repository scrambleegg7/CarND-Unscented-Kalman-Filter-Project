```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rader_file = "nis_radar.txt"
laser_file = "nis_laser.txt"
```

```python
df_rader = pd.read_csv(rader_file,header=None, names=["nis_value"])
df_laser = pd.read_csv(laser_file,header=None, names=["nis_value"])


print(df_rader.shape)
print(df_laser.shape)
```

```python
plt.plot(range( df_rader.shape[0]) , df_rader.values)
plt.axhline(y=7.815, color='#d62728')
```

```python
plt.plot(range( df_laser.shape[0]) , df_laser.values)
plt.axhline(y=7.815, color='#d62728')
```

```python
sns.distplot(df_rader.values)
```

```python
sns.distplot(df_laser)
```
